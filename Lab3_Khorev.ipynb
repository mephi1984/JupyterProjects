````python
2. Построить график произвольной квадратичной функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры
````

````python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_function(x):
    return 2 * x**2 + 3 * x + 1


x = np.linspace(-10, 10, 400)

y = quadratic_function(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = 2x^2 + 3x + 1')

plt.title('График', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)

plt.grid(True)
plt.legend()

plt.show()
````

````python
3. Построить график произвольной гармонической функции, масштаб задать произвольно. Сделать настройки подписей осей, заголовка фигуры

````

````python
import numpy as np
import matplotlib.pyplot as plt


def harmonic_function(x):
    return np.sin(2 * np.pi * x) 


x = np.linspace(0, 4, 400)


y = harmonic_function(x)

plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = sin(2*pi*x)')

plt.title('График', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)

plt.grid(True)
plt.legend()

plt.show()
````

````python
4.	Вывести графики п.2 и 3 в разные поля при помощи функции subplot()
````

````python
import numpy as np
import matplotlib.pyplot as plt

def quadratic_function(x):
    return 2 * x**2 + 3 * x + 1

def harmonic_function(x):
    return np.sin(2 * np.pi * x)

fig, axs = plt.subplots(2, 1, figsize=(8, 10))

x_quadratic = np.linspace(-10, 10, 400)
y_quadratic = quadratic_function(x_quadratic)
axs[0].plot(x_quadratic, y_quadratic, label='y = 2*x*x + 3*x + 1')
axs[0].set_title('График 1', fontsize=16)
axs[0].set_xlabel('X', fontsize=12)
axs[0].set_ylabel('Y', fontsize=12)
axs[0].grid(True)
axs[0].legend()

# Второй график - гармоническая функция
x_harmonic = np.linspace(0, 4, 400)
y_harmonic = harmonic_function(x_harmonic)
axs[1].plot(x_harmonic, y_harmonic, label='y = sin(2*pi*x)')
axs[1].set_title('График 2', fontsize=16)
axs[1].set_xlabel('X', fontsize=12)
axs[1].set_ylabel('Y', fontsize=12)
axs[1].grid(True)
axs[1].legend()

# Отображаем оба графика
plt.tight_layout()
plt.show()
````

````python
5.	Построить графики п.2 и 3 в одной координатной плоскости, сделать подписи графиков, указать легенду.
````

````python

x_quadratic = np.linspace(-10, 10, 400)
y_quadratic = quadratic_function(x_quadratic)

x_harmonic = np.linspace(0, 4, 400)
y_harmonic = harmonic_function(x_harmonic)

plt.figure(figsize=(8, 6))

plt.plot(x_quadratic, y_quadratic, label='y = 2*x*x + 3*x + 1')

plt.plot(x_harmonic, y_harmonic, label='y = sin(2*pi*x)', linestyle='--')

plt.title('График', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)

plt.legend()

plt.grid(True)


plt.show()
````

````python

6.	Построить график функции, заданной параметрически:

`x(t) = cos(3*t)` и `y=sin(2*t)`, `t` принадлежит `[0 .. 2*pi]`
````

````python

t = np.linspace(0, 2 * np.pi, 400)  # t от 0 до 2π
x_parametric = np.cos(3 * t)  # x(t) = cos(3t)
y_parametric = np.sin(2 * t)  # y(t) = sin(2t)


plt.figure(figsize=(8, 6))
plt.plot(x_parametric, y_parametric)

plt.title('x(t) = cos(3*t), y(t) = sin(2*t)', fontsize=16)
plt.xlabel('X', fontsize=12)
plt.ylabel('Y', fontsize=12)

plt.grid(True)

plt.show()
````

````python
7.	Построить столбиковую диаграмму по данным статистического комитета КР по прожиточному минимуму, представленному в таблице
```
Год:	2012	2013	2014	2015	2016	2017	2018	2019	2020	2021	2022
Прожиточный минимум:	4341.15	4599.21	4981.51	5182.99	4794.34	4900.79	4792.54	4806.32	5358.53	6268.31	7178.32
```
````

````python

years = [2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
living_wage = [4341.15, 4599.21, 4981.51, 5182.99, 4794.34, 4900.79, 4792.54, 4806.32, 5358.53, 6268.31, 7178.32]

plt.figure(figsize=(10, 6))
plt.bar(years, living_wage, color='skyblue')

plt.title('Диаграмма', fontsize=16)
plt.xlabel('Год', fontsize=12)
plt.ylabel('Прожиточный минимум', fontsize=12)

plt.grid(True, axis='y', linestyle='--', alpha=0.7)

plt.show()
````

````python
8.	Построить круговой график по данным п. 7
````

````python
plt.figure(figsize=(8, 8))
plt.pie(living_wage, labels=years, autopct='%1.1f%%', startangle=90, colors=plt.cm.Paired.colors)

plt.title('Круговая диаграмма', fontsize=16)

plt.show()
````

````python
9.	Построить график произвольной каркасной поверхности.
````

````python
from mpl_toolkits.mplot3d import Axes3D

x = np.linspace(-5, 5, 400)
y = np.linspace(-5, 5, 400)
x, y = np.meshgrid(x, y)
z = (x**2 - y**2)

fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x, y, z, cmap='viridis')

ax.set_title('Гиперболический параболоид', fontsize=16)
ax.set_xlabel('X', fontsize=12)
ax.set_ylabel('Y', fontsize=12)
ax.set_zlabel('Z', fontsize=12)

plt.show()
````
