```python
1.	Загрузите датасеты классификации из библиотеки skikit-learn - load_cancer
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

cancer = load_breast_cancer()

cancer_df = pd.DataFrame(data=cancer.data, columns=cancer.feature_names)
cancer_df['target'] = cancer.target
```

```python
2.	Изучите данные
```

```python
print(cancer.keys())
```

```python

print("Data:")
print(cancer.data[:5]) 
print("\nFeature Names:")
print(cancer.feature_names) 

print("\nTarget Names:")
print(cancer.target_names) 

print("\nDescription:")
print(cancer.DESCR[:500])  
```

```python
3.	Разделите данные на обучающий и тестовые наборы
```

```python
X = cancer_df.drop('target', axis=1)
y = cancer_df['target']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=55)
```

```python
4.	Исследуйте данные с помощью визуализации. Подберите нужные параметры графиков
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='target', data=cancer_df, palette='viridis')
plt.title('Distribution of Breast Cancer Classes')
plt.xlabel('Cancer Class')
plt.ylabel('Count')
plt.xticks([0, 1], ['Malignant', 'Benign'])
plt.show()
```

```python
selected_features = ['mean radius', 'mean texture', 'mean perimeter', 'mean area']

plt.figure(figsize=(12, 10))

sns.scatterplot(data=cancer_df, x='mean radius', y='mean texture', hue='target', palette='viridis')
plt.title('Scatterplot of Mean Radius vs Mean Texture')
plt.xlabel('Mean Radius')
plt.ylabel('Mean Texture')
plt.legend(title='Cancer Class', loc='upper right', labels=['Malignant', 'Benign'])
plt.show()
```

```python
plt.figure(figsize=(12, 8))
correlation_matrix = cancer_df.corr()
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', cmap='coolwarm', square=True, cbar=True)
plt.title('Correlation Matrix of Breast Cancer Features')
plt.show()
```

```python
5.	Постройте модель по методу k-ближайших средних. Протестируйте на произвольном наборе данных
```

```python
knn = KNeighborsClassifier(n_neighbors=5) 
knn.fit(X_train, y_train)

random_data = np.random.uniform(low=X.min(), high=X.max(), size=(1, X.shape[1]))
predicted_class = knn.predict(random_data)

print(f'Random data: {random_data}')
print(f'Predicted class for random data: {predicted_class[0]}')
```

```python
6.	Получите прогноз модели
```

```python
y_pred = knn.predict(X_test)

```

```python
7.	Оцените качество модели
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy on test set: {accuracy:.2f}')
```

```python
Вывод:

Точность модели k-ближайших соседей в 0.94 указывает на высокую эффективность в классификации доброкачественных
и злокачественных опухолей в наборе данных о раке груди.
Это значение свидетельствует о том, что модель успешно распознает классы, основываясь на предоставленных признаках. 
```
