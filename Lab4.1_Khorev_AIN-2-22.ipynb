```python
1.	Загрузите датасеты классификации из библиотеки skikit-learn - load_digits
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

digits = load_digits()
```

```python
2.	Изучите данные
```

```python
print(digits.keys())
```

```python

print("Data:")
print(digits.data[:5]) 
print("\nFeature Names:")
print(digits.feature_names) 

print("\nTarget Names:")
print(digits.target_names) 

print("\nDescription:")
print(digits.DESCR[:500])  
```

```python
3.	Разделите данные на обучающий и тестовые наборы
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    digits.data, digits.target, test_size=0.25, random_state=66)
```

```python
4.	Исследуйте данные с помощью визуализации. Подберите нужные параметры графиков
```

```python


fig, axes = plt.subplots(2, 5, figsize=(10, 5))
for ax, image, label in zip(axes.ravel(), digits.images, digits.target):
    ax.imshow(image, cmap='gray')
    ax.set_title(f'Label: {label}')
    ax.axis('off')
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 6))
sns.countplot(x=digits.target, palette='viridis')
plt.title('Distribution of Digits')
plt.xlabel('Digit')
plt.ylabel('Count')
plt.xticks(range(10))  # Установка меток оси x от 0 до 9
plt.show()
```

```python
5.	Постройте модель по методу k-ближайших средних. Протестируйте на произвольном наборе данных
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

X = digits.data
y = digits.target

random_data = np.random.rand(1, X.shape[1]) * (X.max() - X.min()) + X.min()
predicted_class = knn.predict(random_data)

print(f'Random data: {random_data}')
print(f'Predicted class for random data: {predicted_class[0]}')
```

```python
6.	Получите прогноз модели
```

```python
y_pred = knn.predict(X_test)
```

```python
7.	Оцените качество модели
```

```python
accuracy = accuracy_score(y_test, y_pred)

print(f'Accuracy: {accuracy:.2f}')
```

```python
Вывод:
Точность 0.99 указывает на то, что модель хорошо справляется с задачей классификации рукописных цифр в наборе данных digits.
Это высокий уровень точности, что может свидетельствовать о том,
что модель эффективно извлекает и использует информацию из признаков для различения цифр. 
```
